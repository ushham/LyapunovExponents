## Defining the system
The DynamicalSystem library requires the user to provide the `time`, the `variables`, and the `parameters` as sympy symbols.

Using these symbols the system equations are provided as a list.

All of these, along with the optional numerical values for the parameters, are then passed into the DynamicalSystem class.

```python
from sympy import symbols
import sympy as sy

#Example of implementing new dynamical system
t = symbols('t')
x, y = symbols('x, y')
beta, alpha, q, eta = symbols('beta, alpha, q, eta')

variables = [x, y]

parameters = [beta, alpha, q, eta]
parms=[0.05, 1, 0.9, 0.5]

duffing_system = [
    y,
    -beta * y - alpha * x ** 3 + q * sy.cos(eta * t)
]

duffing = DynamicalSystem(time=t, var=variables, params=parameters, item=duffing_system)
duffing.update_parameters(parms)
duffing.print_system()
```

## Defining the Lyapunov class
The calculations of the Lyapunov exponents and vectors are completed by the `Lyapunov` class, by passing the system into the class.

```python
lyap = Lyapunov(duffing)
```

```python
fwd = lyap.forwards
bkw = lyap.backwards

bkw.set_params({'max_time': 500})
bkw.set_params({'max_time': 500})

fwd._set_data()
bkw._set_data()
```

```python
np.set_printoptions(precision=4, suppress=True)
print(fwd.exponents_average())
print(bkw.exponents_average())
```

```python
lyap.plot_trajectory()
```
